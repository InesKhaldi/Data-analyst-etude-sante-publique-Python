```python
#Transformation du type des variables 'Zone' en string et 'Année' en integer
df_sous_nutrition['Zone'] = df_sous_nutrition['Zone'].astype('string')
df_sous_nutrition['Année'] = df_sous_nutrition['Année'].astype(int)
```

    3.1 La proportion de personnes en sous-nutrition en 2017  <a class='anchor' id='3.1-la-proportion-de-personnes-en-sous-nutrition-en-2017'></a>
    </h3>

**Réalisation d'une jointure pour créer un nouveau df filtré sur l'année 2017**

```python
#Création de deux dataframe correspondant à l'année 2017
df_population_2017 = df_population.loc[df_population['Année']== 2017]
df_sous_nutrition_2017 = df_sous_nutrition.loc[df_sous_nutrition['Année']== 2017]

#Réalisation d'une jointure entre les deux df
df_pop_sous_nutrition = pd.merge(df_population_2017, df_sous_nutrition_2017, on='Zone', how='inner')
```

```python
#Afficher le dataset
df_pop_sous_nutrition
```

```python
#Supprimer les colonnes 'Année_y' et 'Année_x' et afficher le dataset
del df_pop_sous_nutrition['Année_y']
del df_pop_sous_nutrition['Année_x']
df_pop_sous_nutrition
```

---

**Calcul de la population mondiale et la population en sous-nutrition**

```python
#Affichage de la population mondiale en 2017
pop_mondiale = df_population_2017['Population'].sum()
print('La population mondiale en 2017 était de', '{:,}'.format(pop_mondiale).replace(',', ' '), 'habitants.')

#Affichage du nombre de personnes sous-nutries en 2017
pers_sous_nutries = df_pop_sous_nutrition['Population en sous-nutrition'].sum()
print ('Il y avait', '{:,}'.format(pers_sous_nutries).replace(',', ' '), 'de personnes sous-nutries en 2017.') 

 
#Calcul de la proportion de la population en état de sous-nutrition 
prop_pop_sousnutrie = (pers_sous_nutries / pop_mondiale) * 100
print('La proportion de la population en sous-nutrition en 2017 est de', round(prop_pop_sousnutrie, 2),'%.')
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">
    3.2 Le nombre théorique de personnes pouvant être nourries avec la disponibilité alimentaire mondiale  <a class='anchor' id='3.2-le-nombre-theorique-de-personnes-pouvant-etre-nourries-avec-la-disponibilite-alimentaire-mondiale'></a> </h3>

**Réalisation d'une jointure pour créer un nouveau df regroupant la disponibilité alimentaire et la population**

```python
#Jointure entre les df population et dispo_alimentaire
df_dispo_alimentaire_monde = pd.merge(df_population_2017, df_dispo_alimentaire, on='Zone', how='inner')

#Afficher le nouveau dataframe 
df_dispo_alimentaire_monde
```

---

**Calcul de la disponibilité alimentaire en kilocalories**

```python
#Calcul des kcal disponibles mondialement sur une année et création d'une colonne
df_dispo_alimentaire_monde['Dispo alimentaire totale en Kcal'] = df_dispo_alimentaire_monde['Disponibilité alimentaire (Kcal/personne/jour)'] * df_dispo_alimentaire_monde['Population'] * 365

print('Le nombre de kcal mondial est de', '{:,}'.format(int(df_dispo_alimentaire_monde['Dispo alimentaire totale en Kcal'].sum())).replace(',', ' '), 'Kcal.')
```

> La quantité d'aliments disponible par habitant est de 2500 kilocalories par jour, seuil fixé par la FAO. 

```python
#Calcul du nombre de personnes pouvant être nourries : disponibilité alimentaire mondiale totale /(nombre de kcal d'une personne/jour * 365)
nb_pers_nourries = df_dispo_alimentaire_monde['Dispo alimentaire totale en Kcal'].sum() / (2500*365)
print('Théoriquement,', '{:,}'.format(int(nb_pers_nourries)).replace(',', ' '), 'personnes peuvent être nourries avec cette disponibilité alimentaire mondiale en kilocalories.')
print('Soit', round(nb_pers_nourries * 100 / pop_mondiale, 2), '% de la population en 2017.')
```

> En terme de kilocalories, la disponibilité alimentaire mondiale permet de nourrir plus de la totalité de la population en quantité et qualité suffisante.

**Création d'un nouveau df filtré sur les aliments d'origine végétale et réalisation d'une jointure**

```python
#Création d'un nouveau df 
df_dispo_alim_vege = df_dispo_alimentaire.loc[df_dispo_alimentaire['Origine']== 'vegetale']
df_dispo_alim_vege.head()
```

```python
#Jointure entre les df populaltion et dispo_alim_vege
df_dispo_alim_vege_pop = pd.merge(df_population_2017, df_dispo_alim_vege, on='Zone', how='inner')

#Afficher le nouveau dataframe 
df_dispo_alim_vege_pop
```

---

**Calculs**

```python
#Calcul du nombre kcal disponibles pour les végétaux
df_dispo_alim_vege_pop['Dispo alimentaire végétaux en Kcal'] = df_dispo_alim_vege_pop['Disponibilité alimentaire (Kcal/personne/jour)'] * df_dispo_alim_vege_pop['Population'] * 365

print('Le nombre de kcal disponibles des produits végétaux  est de','{:,}'.format(int(df_dispo_alim_vege_pop['Dispo alimentaire végétaux en Kcal'].sum())).replace(',', ' '), 'Kcal.')
```

```python
#La quantité d'aliments disponible par habitant est de 2500 kilocalories par jour, le seuil fixée par la FAO.

#Calcul du nombre de personnes pouvant être nourries avec les végétaux
nb_pers_nourries_vegetaux = df_dispo_alim_vege_pop['Dispo alimentaire végétaux en Kcal'].sum() / (2500*365)
print('Théoriquement,', '{:,}'.format(int(nb_pers_nourries_vegetaux)).replace(',', ' '), 'personnes peuvent être nourries avec cette disponibilité alimentaire mondiale en kilocalories.')
print('Soit', round(nb_pers_nourries_vegetaux * 100 / pop_mondiale, 2), '% de la population en 2017.')
```

> Près de la totalité de la population pourrait se nourrir avec seulement des produits végétaux.

<h4 style="margin: auto; padding: 20px; color: RGB(0,114,200)">
    4.1.1 Vérification de la relation de la disponibilité intérieure  <a class='anchor' id='4.1.1-verification-de-la-relation-de-la-disponibilite-interieure'></a>
    </h4>

**La disponibilité intérieure correspond à la disponibilité totale d'un aliment au sein du pays.**
***
>Il y a un lien logique entre l'ensemble des 10 variables (milliers de tonnes) sur l'utilisation de la disponibilité intérieure. 

>On vérifie donc la relation suivante : Production + Importations - Exportations + Variation de stock = **Disponibilité intérieure** = Semences + Pertes + Nourriture + Aliments pour animaux + Traitement + Autres utilisations

**Vérification de la 1° relation**

```python
#Disponibilité intérieure = Production + Importations - Exportations + Variation de stock  
df_dispo_alimentaire['prod_imp_exp_var']=df_dispo_alimentaire['Production'] + (df_dispo_alimentaire['Importations - Quantité'] - df_dispo_alimentaire['Exportations - Quantité']) + df_dispo_alimentaire['Variation de stock']
df_dispo_alimentaire[['Disponibilité intérieure', 'prod_imp_exp_var']]
```

Cette première relation est vérifiée.

**Vérification de la 2° relation**

```python
#Disponibilité intérieure = Semences + Pertes + Nourriture + Aliments pour animaux + Traitement + Autres utilisations
df_dispo_alimentaire['Utilisation Dispo Alim']= df_dispo_alimentaire['Semences'] + df_dispo_alimentaire['Pertes'] + df_dispo_alimentaire['Nourriture'] + df_dispo_alimentaire['Aliments pour animaux'] + df_dispo_alimentaire['Traitement'] + df_dispo_alimentaire['Autres Utilisations']
df_dispo_alimentaire[['Disponibilité intérieure', 'Utilisation Dispo Alim']]
```

La deuxième relation est vérifiée. 

```python
#Suppression des colonnes crées pour vérifier la relation 
del df_dispo_alimentaire['prod_imp_exp_var']
del df_dispo_alimentaire['Utilisation Dispo Alim']
df_dispo_alimentaire
```

**Le nombre de personnes pouvant théoriquement être nourries avec la quantité de manioc exportée.**
>Le manioc apporte une valeur énergétique de 159 kcal pour 100g. Soit 1590 kcal pour 1 kg de manioc. 

*Source : lanutrition.fr*

```python
#Utilisation de la méthode iloc[0] pour extraire la valeur numérique (permet d'afficher uniquement la valeur calculée sans aucun texte : dtype: float64)
kcal_kg_manioc = 1590
nb_pers_nourries_manioc = df_dispo_thailande_manioc['Exportations - Quantité'].iloc[0]* kcal_kg_manioc / (2500*365) 
print('Théoriquement,', '{:,}'.format(int(nb_pers_nourries_manioc)).replace(',', ' '), 'personnes peuvent être nourries avec la quantité de manioc exportée.')
print('Soit', round(nb_pers_nourries_manioc * 100 / df_dispo_thailande_manioc['Population'].iloc[0],2), '% des Thaïlandais en 2017.')
```
